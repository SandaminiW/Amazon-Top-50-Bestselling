```python
# Group by "Year" and calculate the mean only for numeric columns
p_mean = df.groupby("Year").mean(numeric_only=True)

# Assign the new list of years to the 'Year' column
p_mean['Year'] = [2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]

# Reset the index
p_mean.reset_index(drop=True, inplace=True)

# Display the result
p_mean
```

```python
sns.regplot(x="Year", y="Price",data=p_mean)
```
